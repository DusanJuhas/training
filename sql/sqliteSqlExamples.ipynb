# Introduction to SQL with SQLite
Hands-on examples for beginners with Python background.

## Setup Sample Database

```python
import sqlite3

conn = sqlite3.connect("training.db")
cur = conn.cursor()

# Create tables
cur.execute("""
CREATE TABLE IF NOT EXISTS customers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    city TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS orders (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    amount REAL,
    order_date TEXT,
    FOREIGN KEY(customer_id) REFERENCES customers(id)
)
""")

conn.commit()
```

## Insert Sample Data

```python
cur.execute("DELETE FROM orders")
cur.execute("DELETE FROM customers")

customers = [("Alice", "Prague"), ("Bob", "Brno"), ("Carol", "Prague")]
cur.executemany("INSERT INTO customers(name, city) VALUES (?, ?)", customers)

orders = [(1,120.50,"2025-01-10"),(1,250.00,"2025-02-15"),(2,75.00,"2025-03-01"),(3,300.00,"2025-03-20")]
cur.executemany("INSERT INTO orders(customer_id, amount, order_date) VALUES (?, ?, ?)", orders)
conn.commit()
```

## SELECT

```python
cur.execute("""SELECT id, name, city FROM customers""")
cur.fetchall()
```

## WHERE Filter

```python
cur.execute("""SELECT * FROM customers WHERE city = 'Prague'""")
cur.fetchall()
```

## ORDER BY and LIMIT

```python
cur.execute("""SELECT * FROM orders ORDER BY amount DESC LIMIT 4""")
cur.fetchall()
```

## Aggregations

```python
cur.execute("""SELECT customer_id, SUM(amount) AS total_sales FROM orders GROUP BY customer_id""")
cur.fetchall()
```

## HAVING

```python
cur.execute("""SELECT customer_id, SUM(amount) AS total_sales FROM orders GROUP BY customer_id HAVING SUM(amount) > 100""")
cur.fetchall()
```

## INNER JOIN

```python
cur.execute("""SELECT c.name, c.city, o.amount FROM customers c JOIN orders o ON c.id = o.customer_id""")
cur.fetchall()
```

## LEFT JOIN

A LEFT JOIN is used when you want to return all rows from the left table, even if there is no matching row in the right table.

Why use LEFT JOIN?
 - Find customers who have no orders.
 - Show all employees, including those without assigned projects.
 - Display all products, including those never sold.


```python
cur.execute("""SELECT c.name, o.amount FROM customers c LEFT JOIN orders o ON c.id = o.customer_id""")
cur.fetchall()
```

## UPDATE

```python
cur.execute("""UPDATE customers SET city = 'Plzen' WHERE id = 2""")
conn.commit()
```

## DELETE

```python
cur.execute("""DELETE FROM orders WHERE amount < 100""")
conn.commit()
```
