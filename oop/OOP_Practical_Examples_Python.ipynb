# Practical OOP Examples in Python
Hands-on examples covering classes, objects, encapsulation, inheritance, polymorphism, composition, and a mini project.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def describe(self):
        return f"{self.brand} {self.model}"

car = Car("Skoda", "Octavia")
print(car.describe())
```

## Encapsulation

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance

acc = BankAccount("Alice", 100)
acc.deposit(50)
print(acc.get_balance())
```

## Inheritance

```python
class Employee:
    def __init__(self, name):
        self.name = name

    def work(self):
        print(f"{self.name} is working")

class Developer(Employee):
    def code(self):
        print(f"{self.name} writes Python code")

dev = Developer("John")
dev.work()
dev.code()
```

## Polymorphism

```python
class Dog:
    def speak(self):
        return "Woof"

class Cat:
    def speak(self):
        return "Meow"

for animal in [Dog(), Cat()]:
    print(animal.speak())
```

## Composition

```python
class Engine:
    def start(self):
        return "Engine started"

class Vehicle:
    def __init__(self):
        self.engine = Engine()

car = Vehicle()
print(car.engine.start())
```

## Mini Project: Employee Management System

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculate_bonus(self):
        return self.salary * 0.05

class Manager(Employee):
    def calculate_bonus(self):
        return self.salary * 0.10

staff = [Employee("Anna", 50000), Manager("Peter", 70000)]
for emp in staff:
    print(emp.name, emp.calculate_bonus())
```
